# Search with a predicate function

The most powerful search capability is if you use functions that return True for notebooks that match. Each function should take an NB object. The magic function doesn't work for this. You have to use a function from csearch instead.

```
from csearch import csearchf

def func(NB):
  c1 = 'tag1' in NB.get_tags()
  c2 = NB.search_headings('tag1')
  return c1 and c2

csearchf("My Drive/csearch/csearch/examples",  func)
```
