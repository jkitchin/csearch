# Get links to all notebooks

```
# Note this relies on this directory existing in your drive.
%csearch "/usr/local/lib/python3.6/dist-packages/csearch/examples" -lr
```

# search Markdown

```
%csearch "/usr/local/lib/python3.6/dist-packages/csearch/examples" -r -m independent
```

# Search code

```
%csearch "/usr/local/lib/python3.6/dist-packages/csearch/examples" -r -c independent
```

# search tags

```
%csearch "/usr/local/lib/python3.6/dist-packages/csearch/examples" -r -t tag1
```

```
%csearch "/usr/local/lib/python3.6/dist-packages/csearch/examples" -r -t !tag1 tag2
```

# Search properties

```
%csearch "/usr/local/lib/python3.6/dist-packages/csearch/examples" -r -p author=jkitchin
```

# Search headings

```
%csearch "/usr/local/lib/python3.6/dist-packages/csearch/examples" -r -H tags
```

# Search todo 

```
%csearch "/usr/local/lib/python3.6/dist-packages/csearch/examples" -r -d leading
```

```
%csearch "/usr/local/lib/python3.6/dist-packages/csearch/examples" -r -d due
```

# Combine criteria

## A simple query with just tags gets two results

```
%csearch "/usr/local/lib/python3.6/dist-packages/csearch/examples" -r -t tag1 
```

## Narrow with multiple criteria

```
%csearch "/usr/local/lib/python3.6/dist-packages/csearch/examples" -r -t tag1 -H heading
```
